# Classic Method

## Single Plot

```python
# Let's create some data
import numpy as np
x=np.linspace(0, 2*np.pi,100)
y=np.sin(x)
```

```python
plt.plot(x,y,'r-')
plt.xlabel('Angle (rad)')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
```

## Multiple Plots

```python
# Another set of data
y2=np.cos(x)

plt.subplot(1,2,1) # (Number of Rows, Number of Columns, and the Figure Number)
plt.plot(x,y,'r')

plt.subplot(1,2,2) 
plt.plot(x,y2,'b')
```

```python
# Another Example
y3=x**2
y4=x**0.5

plt.subplot(2,2,1) # Number of Rows, Number of Columns, and the Figure Number
plt.plot(x,y,'r')

plt.subplot(2,2,2) # Number of Rows, Number of Columns, and the Figure Number
plt.plot(x,y2,'b')

plt.subplot(2,2,3) # Number of Rows, Number of Columns, and the Figure Number
plt.plot(x,y3,'g')

plt.subplot(2,2,4) # Number of Rows, Number of Columns, and the Figure Number
plt.plot(x,y4,'m')
```

# Object Oriented Method

## Simple Plot

```python
# Step 1. Create an Object
fig=plt.figure()

# Step 2. Assign Axes to this Object
ax=fig.add_axes([0.1,0.1,1,0.5]) # ([Bottom Left Corner Coordinates, Width, and height of the Plot])

# Step 3. Add data to this axes
ax.plot(x,y)
```

## Inserting a Plot inside a Plot

```python
fig=plt.figure()

ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
# Let's repeat them again----
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])

# Now, you can add Plot to these axes

ax1.plot(x,y3)
ax1.set_title('Result 1')


ax2.plot(x,y4,'r')
ax2.set_title('Other Result')
```

## Multiple Plots

```python
fig,ax=plt.subplots(nrows=1, ncols=2)
```

```python
fig,ax=plt.subplots(nrows=2, ncols=2)

plt.tight_layout() # This is just to get clean and neat style
```

```python
fig,ax=plt.subplots(nrows=1, ncols=2)

ax[0].plot(x,y)
ax[1].plot(x,y3,'m--')
```

```python
fig,ax=plt.subplots(nrows=2, ncols=2)
plt.tight_layout()

ax[0,0].plot(x,y)
ax[0,1].plot(y,x,'r--')
ax[1,0].plot(y2,x)
ax[1,1].plot(x,y3)
```

# Controllign the Size of a Figure

```python
fig=plt.figure(figsize=(3,2), dpi=100)  # DPI=Dot Per Inch

ax=fig.add_axes([0,0,1,2])

ax.plot(x,y)
```

# Save a Figure

```python
# Let's Create a Figure
fig=plt.figure(figsize=(3,2), dpi=100)  # DPI=Dot Per Inch
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y, 'r--')
ax.set_title('A Sine Wave')
ax.set_xlabel('X')
ax.set_ylabel('Y')

fig.savefig('Files\My_Plot.jpg')
```

# Add Legend

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be placed. The allowed values of **loc** are numerical codes for the various places the legend can be placed. See the [documentation page](https://matplotlib.org/stable/api/legend_api.html?highlight=legend#module-matplotlib.legend) for details. 

Or you can specify the coordinate of the corner of the legend, by for example loc=(0.1,0.2)

```python
fig=plt.figure(figsize=(3,2), dpi=100)  # DPI=Dot Per Inch, Note: The size is in Inch
ax=fig.add_axes([0,0,2,2])

ax.plot(x,y, 'r--', label='Sine')
ax.plot(x,y2,'m', label='Cosine')

ax.legend(loc=0)  # Location=0, means we let Python decides the best location 


ax.set_title('A Sine & Cosine Wave')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

# Linewidth and Style of a Plot

```python
# Some random data points 
x=np.linspace(1,3,10)
y=1/x+x**0.5
```

```python
MyFig=plt.figure()
Ax=MyFig.add_axes([0,0,1,1])
Ax.plot(x,y, color='blue', linewidth='0.75', alpha=0.75, linestyle='--',
        marker='o',markersize=20,  markerfacecolor='yellow')  # alpha control the transparency 
```

```python
# Another Example
MyFig=plt.figure()
Ax=MyFig.add_axes([0,0,1,1])
Ax.plot(x,y, color='red', linewidth='0.75', alpha=0.75, linestyle='--',
        marker='o',markersize=20,  markerfacecolor='yellow', markeredgewidth=2, markeredgecolor='green')  # alpha control the transparency 
```

```python
MyFig=plt.figure()
Ax=MyFig.add_axes([0,0,1,1])
Ax.plot(x,y, color='purple', linewidth='1', alpha=0.75, linestyle='--',
        marker='1',markersize=20, markerfacecolor='yellow')  # alpha control the transparency 
```

# Set the X and Y limit

```python
MyFig=plt.figure()
Ax=MyFig.add_axes([0,0,1,1])
Ax.plot(x,y)

Ax.set_xlim([1, 2])
Ax.set_ylim([1.8,2.0])
```

# Histogram

```python
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data, bins=20)
```

# Box Plot

```python
# Let's create 3 arrays of 100 normal random data points with standard Deviation of 1, 2, and 3
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
np.shape(data)
```

```python
# We can do a quick plot to see the data
plt.plot(data[0],'r')
plt.plot(data[1],'b')
plt.plot(data[2],'m')
```

```python
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True); 
```
