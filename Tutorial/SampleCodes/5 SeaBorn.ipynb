# Import Library

```python
import seaborn as sb
```

# Data
We can use some of the built-in data in this library as example to start with. Let's grab the dataset called, 'Tips'

```python
Tips=sb.load_dataset('tips')
```

```python
Tips
```

# A Quick Check of the Data

```python
sb.histplot(Tips['tip'])
```

```python
sb.histplot(Tips['total_bill'], kde=True)  # kde fit a Distribution Function to the Data
```

```python
# Change the number of bins
sb.histplot(Tips['total_bill'], kde=True, bins=30)
```

# The PDF (Probability Density Function)
If interested in learning more, PDF is built here based on KDE [Kernel density estimation](https://en.wikipedia.org/wiki/Kernel_density_estimation#:~:text=In%20statistics%2C%20kernel%20density%20estimation,based%20on%20kernels%20as%20weights.)

```python
sb.kdeplot(Tips['total_bill'])
```

# Joint Plot

<font color='red'>
<font size='4'>
Please note, in future we will learn how to interpret and use each of these plots for different ML; but here, you just need to learn how to create each of these plots.

```python
sb.jointplot(x='total_bill' , y='tip' , data=Tips)
```

```python
# In Future, we will learn, how we can check the trend of data, for example, here how Tip and Total_Bill are related
sb.jointplot(x='total_bill' , y='tip' , data=Tips, kind='reg')
```

```python
# We can look at the density or a 2D Distribution (future discussion,  kernel density estimation (KDE))
sb.jointplot(x='total_bill' , y='tip' , data=Tips, kind='kde')
```

# Pair Plots (a Quick Way to Visualize the Data)
By comparing each pair of data <br>
<font size='4'>
<font color='magenta'>
Note: depending on Speed of your computer, it may takes 3-10 seconds to produce this plot. 

```python
sb.pairplot(Tips)
```

```python
# You may visualize the data based on Categorical Data; For instance, Male & female
sb.pairplot(Tips, hue='sex')
```

# categorical data

```python
sb.countplot(x='sex',data=Tips)
```

```python
sb.barplot(x='sex',y='total_bill', data=Tips)
```

# Box & Whisker Plot

# Violin Plot

```python
sb.violinplot(x='day', y='tip', data=Tips, hue='sex', split=True)
```

# Matrix Plot (Heat Map)
Illustrates the corrolation between variables (we can actually get the values; but it is also good to have a visual)

```python
# To get the values (i.e. how each pair of variables are correlated )
Tips.corr()
```

```python
sb.heatmap(Tips.corr(), annot=True)
```

# Control The Style of Plot

```python
sb.set_style('whitegrid')
# OR
# sb.set_style('ticks')

sb.barplot(x='day',y='tip', data=Tips)
```

```python
sb.set_style('ticks')
sb.barplot(x='day',y='tip', data=Tips)

# Now, To remove the Right and Top Spine
sb.despine()
```

```python
# To Control the Size
sb.set(rc={'figure.figsize':(8,6)})

# To Control the Font Size
sb.set_context('poster',font_scale=1)
sb.barplot(x='day',y='tip', data=Tips)

# Shift +Tab to see other options here 
#sb.set_context()
```
