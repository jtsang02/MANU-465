# Frequency Spectrogram

## Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Import the Signal

```python
x=np.loadtxt("mic.txt",unpack=True)
```

```python
plt.plot(np.arange(len(x)), x)
```

## STFT

```python
SamplingFreq=10000    # Sampling Frequency
WindowSize = 256      # the length of the windowing segments  1024   256
OverLap=200
Pxx, freqs, bins, im = plt.specgram(x, NFFT=WindowSize, Fs=SamplingFreq, noverlap=OverLap)
```
